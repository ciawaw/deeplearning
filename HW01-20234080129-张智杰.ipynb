```python
# 4.2 编程题：手动实现简单计算图的反向传播

# 1. 前向计算：输入 x，经过 a = x·w1，b = a + w2，输出 L = b²
x = 2.0
w1 = 1.5
w2 = 0.5

# 前向传播
a = x * w1
b = a + w2
L = b ** 2
print("前向计算结果：")
print(f"a = {a}, b = {b}, L = {L}")

# 2. 手动计算梯度（链式法则）
# ∂L/∂b = 2b
# ∂L/∂w2 = ∂L/∂b * ∂b/∂w2 = 2b * 1 = 2b
# ∂L/∂a = ∂L/∂b * ∂b/∂a = 2b * 1 = 2b
# ∂L/∂w1 = ∂L/∂a * ∂a/∂w1 = 2b * x = 2b·x
grad_L_w1_manual = 2 * b * x
grad_L_w2_manual = 2 * b
print("\n手动计算梯度：")
print(f"∂L/∂w1 = {grad_L_w1_manual}")
print(f"∂L/∂w2 = {grad_L_w2_manual}")

# 3. 使用 NumPy 实现简单的自动微分验证
# 定义变量
x_np = 2.0
w1_np = 1.5
w2_np = 0.5

# 前向传播（保留中间值）
a_np = x_np * w1_np
b_np = a_np + w2_np
L_np = b_np ** 2

# 反向传播（手动实现链式法则，模拟自动微分）
dL_db = 2 * b_np
dL_dw2 = dL_db * 1  # ∂b/∂w2 = 1
dL_da = dL_db * 1    # ∂b/∂a = 1
dL_dw1 = dL_da * x_np  # ∂a/∂w1 = x

print("\nNumPy 手动反向传播结果：")
print(f"∂L/∂w1 = {dL_dw1}")
print(f"∂L/∂w2 = {dL_dw2}")
```
